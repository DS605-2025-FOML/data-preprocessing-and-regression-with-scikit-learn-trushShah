2. Removing egregious outliers

```python
def remove_outliers_iqr(df, column):
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    # condition for selecting those rows
    df_clean = df[(df[column] >= lower_bound) & (df[column] <= upper_bound)]
    return df_clean
```

```python
df_clean = df_missing.copy()

for col in ['price', 'sqft_living', 'bedrooms']:
    df_clean = remove_outliers_iqr(df_clean, col)

print(f"Shape before removing outliers: {df_missing.shape}")
print(f"Shape after removing outliers: {df_clean.shape}") # outliers are now gone
```

JUSTIFICATION

To prevent extreme values from skewing our regression model, we removed outliers using the above suggested 1.5×IQR rule.

Price: Some houses are like having extremely expensive (>$7–8M) cost, which are not typical and would distort predictions.

sqft_living: Very large mansions (>10k sqft) are rare and they are not representative.

Bedrooms: Values like 33 bedrooms are not plausible for King County; this error must be like data entry errors or anomalies.

Removing these points will ensure a more robust model without discarding normal variation.

Que.2: Advanced Feature Engineering

Date_based_feature

```python
# let us check 'date' is a datetime object or not
df_clean['date'] = pd.to_datetime(df_clean['date'])

# Extracting year and month of sale
df_clean['sale_year'] = df_clean['date'].dt.year
df_clean['sale_month'] = df_clean['date'].dt.month

# Calculating the age of house at the time of sale
df_clean['house_age'] = df_clean['sale_year'] - df_clean['yr_built']

df_clean[['date', 'sale_year', 'sale_month', 'yr_built', 'house_age']].head()
```

Renovation Status

```python
# Creating the mentioned binary feature (was_renovated)
df_clean['was_renovated'] = df_clean['yr_renovated'].apply(lambda x: 1 if x != 0 else 0)
```

```python
# creating another feature (age_since_renovation)
def compute_age_since_renovation(row):
    if row['was_renovated'] == 1:
        return row['sale_year'] - row['yr_renovated']
    else:
        return row['house_age']

df_clean['age_since_renovation'] = df_clean.apply(compute_age_since_renovation, axis=1)
```

```python
# let us see, if they are made or not
df_clean[['yr_renovated', 'was_renovated', 'house_age', 'age_since_renovation']].head(10)
```

Ratio_features

```python
# first feature (sqft_living_per_floor)
# to avoid division by zero, we can replace 0 floors with 1 temporarily.
df_clean['floors_safe'] = df_clean['floors'].replace(0, 1)

df_clean['sqft_living_per_floor'] = df_clean['sqft_living'] / df_clean['floors_safe']

# Drop temporary column
df_clean.drop(columns=['floors_safe'], inplace=True)
```

```python
# second ratio feature (bath_per_bed)
# to avoid division by zero, we can replace 0 bedrooms with 1 temporarily
df_clean['bedrooms_safe'] = df_clean['bedrooms'].replace(0, 1)

df_clean['bath_per_bed'] = df_clean['bathrooms'] / df_clean['bedrooms_safe']

# Drop temporary column
df_clean.drop(columns=['bedrooms_safe'], inplace=True)
```

TASK_2: Build a Sophisticated Preprocessing Pipeline

1. Identify Feature Types

```python
# lets identify our feature
# predicting this variable is our main goal
target = "price"

numeric_features = [
    "sqft_living",
    "sqft_lot",
    "bedrooms",
    "bathrooms",
    "floors",
    "sqft_living_per_floor",
    "bath_per_bed",
    "house_age",
    "age_since_renovation"
]

categorical_features = [
    "zipcode",
    "waterfront",
    "view"
]

passthrough_features = [
    "grade",
    "condition",
    "was_renovated",
    "sale_year",
    "sale_month"
]

# these are our overall features.
```

2. Create Preprocessing Pipeline for Each Type

```python
pip install scikit-learn
```

```python
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.impute import SimpleImputer

# for creating numeric pipeline
numeric_pipeline = Pipeline(steps=[
    ("imputer", SimpleImputer(strategy="median")),
    ("scaler", StandardScaler())
])
```

```python
# for creating categorical pipeline
from sklearn.preprocessing import OneHotEncoder

categorical_pipeline = Pipeline(steps=[
    ("encoder", OneHotEncoder(handle_unknown="ignore"))
])
```

3. Combine Pipeline with ColumnTransformer

```python
from sklearn.compose import ColumnTransformer

# Building the ColumnTransformer
preprocessor = ColumnTransformer(
    transformers=[
        ("num", numeric_pipeline, numeric_features),
        ("cat", categorical_pipeline, categorical_features),
        ("passthrough", "passthrough", passthrough_features)
    ]
)
```

```python
from sklearn.pipeline import Pipeline
from sklearn.ensemble import RandomForestRegressor, GradientBoostingRegressor
from xgboost import XGBRegressor  

# 1. Random Forest pipeline
rf_pipeline = Pipeline(steps=[
    ("preprocessor", preprocessor),
    ("model", RandomForestRegressor(random_state=42))
])

# 2. Gradient Boosting pipeline
gb_pipeline = Pipeline(steps=[
    ("preprocessor", preprocessor),
    ("model", GradientBoostingRegressor(random_state=42))
])

# 3. XGBoost pipeline
xgb_pipeline = Pipeline(steps=[
    ("preprocessor", preprocessor),
    ("model", XGBRegressor(random_state=42, verbosity=0))
])
```

2. Hyperparameter tuning with GrideSeachCV

```python
X = df.drop(columns=['price']).copy() 

X['was_renovated'] = (X['yr_renovated'] > 0).astype(int)

X['sale_year'] = pd.to_datetime(df['date']).dt.year
X['house_age'] = X['sale_year'] - X['yr_built']

X['age_since_renovation'] = np.where(X['yr_renovated'] > 0,
                                     X['sale_year'] - X['yr_renovated'],
                                     X['house_age'])

X['sqft_living_per_floor'] = X['sqft_living'] / X['floors'].replace(0, 1)

X['bath_per_bed'] = X['bathrooms'] / X['bedrooms'].replace(0, 1)

X['sale_month'] = pd.to_datetime(df['date']).dt.month
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.ensemble import RandomForestRegressor, GradientBoostingRegressor
from sklearn.pipeline import Pipeline
import xgboost as xgb

# Make sure X and y are defined properly
X = df_clean.drop(columns=['price'])   # drop target
y = df_clean['price']

# 1. Random Forest Pipeline
rf_pipeline = Pipeline(steps=[
    ('preprocessor', preprocessor),
    ('regressor', RandomForestRegressor(random_state=42))
])

rf_param_grid = {
    'regressor__n_estimators': [100, 200],
    'regressor__max_depth': [None, 10, 20]
}

rf_grid = GridSearchCV(rf_pipeline, rf_param_grid, cv=3, n_jobs=-1, scoring='r2')


# 2. Gradient Boosting Pipeline
gb_pipeline = Pipeline(steps=[
    ('preprocessor', preprocessor),
    ('regressor', GradientBoostingRegressor(random_state=42))
])

gb_param_grid = {
    'regressor__n_estimators': [100, 200],
    'regressor__learning_rate': [0.05, 0.1],
    'regressor__max_depth': [3, 5]
}

gb_grid = GridSearchCV(gb_pipeline, gb_param_grid, cv=3, n_jobs=-1, scoring='r2')


# 3. XGBoost Pipeline
xgb_pipeline = Pipeline(steps=[
    ('preprocessor', preprocessor),
    ('regressor', xgb.XGBRegressor(random_state=42, verbosity=0))
])

xgb_param_grid = {
    'regressor__n_estimators': [100, 200],
    'regressor__learning_rate': [0.05, 0.1],
    'regressor__max_depth': [3, 5]
}

xgb_grid = GridSearchCV(xgb_pipeline, xgb_param_grid, cv=3, n_jobs=-1, scoring='r2')


# Training all three models
rf_grid.fit(X, y)
gb_grid.fit(X, y)
xgb_grid.fit(X, y)

# Compare results
print("Random Forest best score:", rf_grid.best_score_)
print("Random Forest best params:", rf_grid.best_params_)

print("\nGradient Boosting best score:", gb_grid.best_score_)
print("Gradient Boosting best params:", gb_grid.best_params_)

print("\nXGBoost best score:", xgb_grid.best_score_)
print("XGBoost best params:", xgb_grid.best_params_)
```
